```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Create the webdriver without specifying executable_path in Colab
driver = webdriver.Chrome(options=chrome_options)

# Access the URL
url = "https://fbref.com/en/comps/1/shooting/World-Cup-Stats"
driver.get(url)

# If dynamic content needs time to load, you can use time.sleep()
import time
time.sleep(5)  # Adjust the sleep duration as needed

# Get the page source after content is loaded
html = driver.page_source

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
divs = soup.find_all('div', {'class': 'table_container is_setup'})

for index, div in enumerate(divs, 1):
    players = div.find_all('td', {'class': 'left', 'data-stat': 'player'})
    positions = div.find_all('td', {'class': 'center', 'data-stat': 'position'})
    countries = div.find_all('td', {'class': 'left', 'data-stat': 'team'})
    goals = div.find_all('td', {'class': ['right group_start', 'right iz group_start'], 'data-stat': 'goals'})
    total_shots = div.find_all('td', {'class': 'right', 'data-stat': 'shots'})
    shots_on_target = div.find_all('td', {'class': 'right', 'data-stat': 'shots_on_target'})
    average_shot_distance = div.find_all('td', {'class': ['right non_qual iz', 'right non_qual','right'], 'data-stat': 'average_shot_distance'})
    penalty_scored = div.find_all('td', {'class': 'right', 'data-stat': 'pens_made'})
    penalty_attempts = div.find_all('td', {'class': 'right', 'data-stat': 'pens_att'})

```

```python
print(f'Length of players: {len(players)}')
print(f'Length of positions: {len(positions)}')
print(f'Length of countries: {len(countries)}')
print(f'Length of goals: {len(goals)}')
print(f'Length of total_shots: {len(total_shots)}')
print(f'Length of shots_on_target: {len(shots_on_target)}')
print(f'Length of average_shot_distance: {len(average_shot_distance)}')
print(f'Length of penalty_scored: {len(penalty_scored)}')
print(f'Length of penalty_attempts: {len(penalty_attempts)}')
```

```python
import pandas as pd

# Assuming you have already parsed the HTML and have the BeautifulSoup object in 'soup'
# ...

# Create an empty list to store data
data = []

# Iterate through the divs
for div in divs:
    # Extract data for each column
    players_data = [player.get_text() for player in div.find_all('td', {'class': 'left', 'data-stat': 'player'})]
    positions_data = [position.get_text() for position in div.find_all('td', {'class': 'center', 'data-stat': 'position'})]
    countries_data = [country.get_text()[2:] for country in div.find_all('td', {'class': 'left', 'data-stat': 'team'})]  # Remove the first 2 letters
    goals_data = [goal.get_text() for goal in div.find_all('td', {'class': ['right group_start', 'right iz group_start'], 'data-stat': 'goals'})]
    total_shots_data = [shots.get_text() for shots in div.find_all('td', {'class': 'right', 'data-stat': 'shots'})]
    shots_on_target_data = [target.get_text() for target in div.find_all('td', {'class': 'right', 'data-stat': 'shots_on_target'})]
    average_shot_distance_data = [distance.get_text() for distance in div.find_all('td', {'class': ['right non_qual iz', 'right non_qual','right'], 'data-stat': 'average_shot_distance'})]
    penalty_scored_data = [penalty.get_text() for penalty in div.find_all('td', {'class': 'right', 'data-stat': 'pens_made'})]
    penalty_attempts_data = [attempts.get_text() for attempts in div.find_all('td', {'class': 'right', 'data-stat': 'pens_att'})]

    # Combine data for each row
    row_data = list(zip(players_data, positions_data, countries_data, goals_data, total_shots_data, shots_on_target_data, average_shot_distance_data, penalty_scored_data, penalty_attempts_data))

    # Append row data to the main list
    data.extend(row_data)

# Create the DataFrame
columns = ['Player', 'Position', 'Country', 'Goals', 'Total Shots', 'Shots on Target', 'Average Shot Distance', 'Penalty Scored', 'Penalty Attempts']
attack_df = pd.DataFrame(data, columns=columns)

# Display the DataFrame
attack_df.head()
```
