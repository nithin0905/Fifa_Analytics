```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Create the webdriver without specifying executable_path in Colab
driver = webdriver.Chrome(options=chrome_options)

# Access the URL
url = "https://fbref.com/en/comps/1/possession/World-Cup-Stats"
driver.get(url)

# If dynamic content needs time to load, you can use time.sleep()
import time
time.sleep(5)  # Adjust the sleep duration as needed

# Get the page source after content is loaded
html = driver.page_source

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
divs = soup.find_all('div', {'class': 'table_container is_setup'})

for index, div in enumerate(divs, 1):
    touches = div.find_all('td', {'class': ['right group_start','right iz group_start'], 'data-stat': 'touches'})
    touches_def_pen_area = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'touches_def_pen_area'})
    touches_def_3rd = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'touches_def_3rd'})
    touches_mid_3rd = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'touches_mid_3rd'})
    touches_att_3rd = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'touches_att_3rd'})
    touches_att_pen_area = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'touches_att_pen_area'})

    take_ons = div.find_all('td', {'class': ['right group_start','right iz group_start'], 'data-stat': 'take_ons'})
    take_ons_won = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'take_ons_won'})
    take_ons_tackled = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'take_ons_tackled'})
    take_ons_tackled_percent = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'take_ons_tackled_pct'})
    carries = div.find_all('td', {'class': ['right group_start','right iz group_start'], 'data-stat': 'carries'})

    carries_distance = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'carries_distance'})
    carries_progressive_distance = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'carries_progressive_distance'})
    progressive_carries = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'progressive_carries'})
    carries_into_final_third = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'carries_into_final_third'})


    carries_into_penalty_area = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'carries_into_penalty_area'})
    miscontrols = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'miscontrols'})
    dispossessed = div.find_all('td', {'class': ['right iz','right'], 'data-stat': 'dispossessed'})
```

```python
touches_length = len(touches)
touches_def_pen_area_length = len(touches_def_pen_area)
touches_def_3rd_length = len(touches_def_3rd)
touches_mid_3rd_length = len(touches_mid_3rd)
touches_att_3rd_length = len(touches_att_3rd)
touches_att_pen_area_length = len(touches_att_pen_area)

take_ons_length = len(take_ons)
take_ons_won_length = len(take_ons_won)
take_ons_tackled_length = len(take_ons_tackled)
take_ons_tackled_percent_length = len(take_ons_tackled_percent)
carries_length = len(carries)

carries_distance_length = len(carries_distance)
carries_progressive_distance_length = len(carries_progressive_distance)
progressive_carries_length = len(progressive_carries)
carries_into_final_third_length = len(carries_into_final_third)

carries_into_penalty_area_length = len(carries_into_penalty_area)
miscontrols_length = len(miscontrols)
dispossessed_length = len(dispossessed)

# Print the lengths
print("Length of touches:", touches_length)
print("Length of touches_def_pen_area:", touches_def_pen_area_length)
print("Length of touches_def_3rd:", touches_def_3rd_length)
print("Length of touches_mid_3rd:", touches_mid_3rd_length)
print("Length of touches_att_3rd:", touches_att_3rd_length)
print("Length of touches_att_pen_area:", touches_att_pen_area_length)

print("Length of take_ons:", take_ons_length)
print("Length of take_ons_won:", take_ons_won_length)
print("Length of take_ons_tackled:", take_ons_tackled_length)
print("Length of take_ons_tackled_percent:", take_ons_tackled_percent_length)
print("Length of carries:", carries_length)

print("Length of carries_distance:", carries_distance_length)
print("Length of carries_progressive_distance:", carries_progressive_distance_length)
print("Length of progressive_carries:", progressive_carries_length)
print("Length of carries_into_final_third:", carries_into_final_third_length)

print("Length of carries_into_penalty_area:", carries_into_penalty_area_length)
print("Length of miscontrols:", miscontrols_length)
print("Length of dispossessed:", dispossessed_length)
```

```python
import pandas as pd
import numpy as np

# Assuming you have already extracted the data from the HTML and stored them in the respective lists

# Replace empty strings with NaN in relevant columns
take_ons_tackled_percent_values = [float(percent.text.replace(',', '').replace('%', '')) if percent.text.strip() else np.nan for percent in take_ons_tackled_percent]

data = {
    'Touches': [int(touch.text.replace(',', '')) for touch in touches],
    'Touches Def Pen Area': [int(touch.text.replace(',', '')) for touch in touches_def_pen_area],
    'Touches Def 3rd': [int(touch.text.replace(',', '')) for touch in touches_def_3rd],
    'Touches Mid 3rd': [int(touch.text.replace(',', '')) for touch in touches_mid_3rd],
    'Touches Att 3rd': [int(touch.text.replace(',', '')) for touch in touches_att_3rd],
    'Touches Att Pen Area': [int(touch.text.replace(',', '')) for touch in touches_att_pen_area],
    'Take Ons': [int(ons.text.replace(',', '')) for ons in take_ons],
    'Take Ons Won': [int(won.text.replace(',', '')) for won in take_ons_won],
    'Take Ons Tackled': [int(tackled.text.replace(',', '')) for tackled in take_ons_tackled],
    'Take Ons Tackled Percent': take_ons_tackled_percent_values,
    'Carries': [int(carry.text.replace(',', '')) for carry in carries],
    'Carries Distance': [int(distance.text.replace(',', '')) for distance in carries_distance],
    'Carries Progressive Distance': [int(distance.text.replace(',', '')) for distance in carries_progressive_distance],
    'Progressive Carries': [int(carry.text.replace(',', '')) for carry in progressive_carries],
    'Carries Into Final Third': [int(carry.text.replace(',', '')) for carry in carries_into_final_third],
    'Carries Into Penalty Area': [int(carry.text.replace(',', '')) for carry in carries_into_penalty_area],
    'Miscontrols': [int(mis.text.replace(',', '')) for mis in miscontrols],
    'Dispossessed': [int(dis.text.replace(',', '')) for dis in dispossessed],
}

# Create DataFrame
attack_df_2 = pd.DataFrame(data)

# Display the DataFrame
attack_df_2
```
